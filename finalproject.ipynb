### Scatterplot

```python
# import library
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# correlation between friends and followers
np.corrcoef(df_tweet2["friends"], df_tweet2["followers"])[0,1]
```

```python
# build model to predict followers using friends
x0 = np.array(df_tweet2["friends"]).reshape((-1,1))
y0 = np.array(df_tweet2["followers"])
model0 = LinearRegression()
model0.fit(x0, y0)
r2 = model0.score(x0, y0)
intercept = model0.intercept_
slope = model0.coef_
```

```python
pred0 = model0.predict(x0)
print(pred0)
```

```python
# print intercept, slope, and r2
print("Intercept:", intercept)
print("Slope:", slope)
print("R-squared:", r2)
```

```python
# line equation
# followers = 1.22 * friends + 2301.57
```

```python
plt.scatter(df_tweet2["friends"], df_tweet2["followers"], label="Actual Results")
plt.plot(x0, pred0, label = "Predicted Results", color="red", linewidth=2)
legend = plt.legend(loc="best")
plt.title("Scatterplot (merged): Friends vs. Followers")
plt.xlabel("Friends")
plt.ylabel("Followers")
plt.savefig("merg_ff.png")
plt.show()
```
